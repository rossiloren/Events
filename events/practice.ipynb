```python
import requests
import csv

API_KEY = "AIzaSyC4zKyBA9DYzSdC3ZqGvj_w350KJwRLHDI"
CX = "95793f8c6ed12480a"
QUERY = "eventi Ancona 05/04/2025"

URL = f"https://www.googleapis.com/customsearch/v1?q={QUERY}&key={API_KEY}&cx={CX}"

response = requests.get(URL)
data = response.json()

# Apri un file CSV in modalitÃ  scrittura
with open('eventi_ancona_12_04_2025.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Evento', 'Data', 'Location']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # Scrivi l'intestazione
    writer.writeheader()
    
    # Estrai e scrivi i dati nel CSV
    for item in data.get("items", []):
        # Qui facciamo una piccola assunzione: il nome dell'evento Ã¨ nel titolo,
        # la data Ã¨ nella query della ricerca, e la location Ã¨ nel "snippet"
        # Puoi adattare questa parte per estrarre le informazioni piÃ¹ precisamente
        event_name = item['title']
        date = QUERY.split(' ')[-1]  # Data della ricerca
        location = "Ancona"  # Supponiamo che la location sia Ancona

        writer.writerow({'Evento': event_name, 'Data': date, 'Location': location})

print("CSV creato con successo!")
```
